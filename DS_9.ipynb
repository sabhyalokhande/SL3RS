```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df9=sns.load_dataset('titanic')
df9
```

```python
df9.info()
```

```python
df9.describe()
```

```python
df9.isnull().sum()
```

```python
df9['age'] = df9['age'].fillna(int(df9['age'].mean())).astype(int)
df9
```

```python
df9['embarked'] = df9['embarked'].fillna(df9['embarked'].mode()[0])
```

```python
df9['embark_town'] = df9['embark_town'].fillna(df9['embark_town'].mode()[0])
```

```python
df9['deck'] = df9['deck'].astype('object')  # Fill missing values as 'Unknown' with dtype 'object'
df9['deck'] = df9['deck'].fillna('Unknown')  # Fill missing values as 'Unknown' with dtype 'object'
```

```python
df9.isnull().sum()
```

```python
numeric_columns=['survived','pclass','age','sibsp','parch','fare']
plt.figure(figsize=(10, 6))
sns.boxplot(data=df9[numeric_columns])
plt.xticks(rotation=45)
plt.title('Boxplot for Numeric Columns')
plt.show()
```

```python
Q1 = df9[numeric_columns].quantile(0.25)
Q3 = df9[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outlier_mask = (df9[numeric_columns] < lower_bound) | (df9[numeric_columns] > upper_bound)

df9_no_outliers = df9[~outlier_mask.any(axis=1)]

print("Dataset before outlier removal:", df9.shape)
print("Dataset after outlier removal:", df9_no_outliers.shape)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df9_no_outliers[numeric_columns])
plt.xticks(rotation=45)
plt.title('Boxplot for Numeric Columns')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

numeric_columns=['survived','pclass','age','sibsp','parch','fare']
scaler = MinMaxScaler()
df9[numeric_columns] = scaler.fit_transform(df9[numeric_columns])
df9[numeric_columns].head() 
```

```python
sns.boxplot(x='survived', y='age', data=df9[df9['sex'] == 'male'])
plt.title("Age Distribution for Male Passengers", fontsize=14)
plt.xlabel("Survival Status", fontsize=12)
plt.ylabel("Age", fontsize=12)
```

```python
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
sns.boxplot(x='survived', y='age', data=df9[df9['sex'] == 'female'], palette='Set2')
plt.title("Age Distribution for Female Passengers", fontsize=14)
plt.xlabel("Survival Status", fontsize=12)
plt.ylabel("Age", fontsize=12)
```
